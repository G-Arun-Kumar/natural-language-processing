# Cleaning the texts

Cleaning the text is an essential step in NLP. we will make our text as clean as possible. Our texts are mixed with several punctuations. There are **upper case and lower case letters, special characters**. Cleaning texts will ease the learning process of the future machine learning model which we will train to understand the reviews or english and predict if reviews are positive or negative.

**Stop words** - Stop words are the ensemble of words we don't want to include in our reviews after cleaning the text which we know
the words that are not relevant to help predictions.Eg. 'The', 'a', 'and' are all the words. It includes all **pronouns** as well. For eg. 'I', 'we', 'he', 'she', 'it', 'they', 'them', 'those' etc. Since these words doesn't give meaning to our machine learning model or the bag of words that we create, we omit these words before we build the model.

**Stemming** - It is the process of taking only the root of the words that is more sufficient to give the actual meaning. It is nothing but trimming few letters from a word without spoling the actual meaning of the word. By doing this, we will increase the dimension of the **sparse matrix**. For example, let's take the word **'loved'**. After applying stemming process we will just get the output as **love** which will remove the last 2 letters **'e' and 'd'.** But if we see, the meaning of the word, it **never changed**. We will totally remove all the **conjugation of the verbs** just keeping the present tense of a word there by simplying the reviews.

**Sparse matrix** - We will create a sparse matrix when creating the actual bag of words model. This sparse matrix consists of **different words in each column of all our different reviews.** In order to optimize or minimize the dimension(no.of columns) of the sparse matrix, we need to simplify as much of words as we can. If we don't apply stemming process, then for each word we will have different columns that is a kind of duplicate. For example, the word 'loved' without applying stemming process, we will have 'love' in one column and the word 'loved' in another column. This makes the sparse matrix complex and would be redundant with higher dimensions and that's wrong. So, sparse matrix helps to reduce the final dimension of sparse matrix.

**Tokenization** - The process of creating all the columns corresponding to each of the words taken from all the reviews in a sparse matrix is called **tokenization**

```python
"""We will import 're' library  which will help us to simplify the reviews. We have another library which will take care
of stemming.

There is another library called 'NLTK' which is very classic library in NLP which will allow us to download the ensemble
of stop words.

Next, using download method of nltk library, we need to pass keyowrd 'stopwords'. This will get all the stopwords
and later on these stopwords are not included in these non-relevant words in our reviews.

Next, we will import stopwords method from corpus module of nltk library. So there are two things here. The above
statement will help us to download all the stopwords. This code will help us to import those stopwords to our notebook.
Kind of downloading and installing a software in the outer World.

Next, we will import PorterStemmer class from porter submodule of stem module of nltk library. This will help us to apply
stemming of our reviews.

Next, we will create an empty list called corpus. This corpus list is the final list that contains all cleaned reviews
one by one.

Next, we will create a for loop to iterate through all the reviews that includes all our cleaning process. We will take
each review one by one, clean it and then add(append) it in corpus list. And this list is expected by future functions
as reviews. The range is set to 1000 as upper bound since we have 1000 reviews in our dataset.

Next, we create a variable called 'review' which will have all cleaned data which will be passed as a parameter
to corpus list. We will clean each type and update it in each step. First kind of cleaning will be to remove all
punctuations(,...,'',""etc). Simply, we need just alphabets and anything other than that will be removed.
We use 're' library and sub() to execute this. This sub() has 3 parameters.
They are 'pattern'(what will be replaced. Those punctuations).The trick to give pattern is in square brackets ^(hat which
means not. So 'not a-z and A-Z' both lower & upper case. It means anything that is not alphabets).
Second parameter'replacement'(replacement type. A space here in our problem). Third parameter is 'string'(from where
the source of words or text comes. It's our dataset in our example). The way we give is specify dataframe and specify
column name within square bracket also mention [i] in another square bracket. This means since we are in for loop we need
only first review and not all reviews. In the next-next iterations we will get reviews one by one. If we didn't
mention [i] then we will get all reviews which is wrong getting wrong output.
This sub() is similar to regex pattern.

Our next cleaning work is to make all reviews as lower case. Changing all upper case or capital letters to lowercase.
And we use lower() to achieve this and store the result in review variable itself.

Next, we will need to split each review as each words. This is required for stemming process. So we use split()
We merge back all the individual words into a single review(back to original state) after the stemming process
is done. And we use join() to achieve this.

Next, we create an object for PorterStemmer class. This object will be the output of the next for loop which we use.

Next, we create a variable called 'all_stopwords'. We now use words() in stopwords module and pass the parameter
as 'english'. Then it is assigned to the variable 'all_stopwords'.

Next, we remove the 'not' keywords from the ensemble of words in the stopwords. This is because when we first tried
implementing the stemming, the stemming() removed the 'not' word from the review which is wrong. It shouldn't
remove the 'not' keyword since we need negative review in bag of words model. So, we are removing the 'not' keyword
from the ensemble of stopwords, so that if this is passed to the stemming() below, the not words will also be included
for further when building bag of words model to give the correct prediction. Hence, we use remove() of all_stopwords
variable and pass the keyword 'not' as a parameter.

Next, we start the stemming procedure(get only root of the words without changing the meaning). We now use
list comprehension with if condition here to achieve the stemming. applying one more for loop(taking word after word,
apply stemming procedure). Also include only 'non-stopwords' in stemming process.
The trick to do is, use 'if not' condition. It means, accept all words if not stop words. Filter out all stopwords using
if not syntax. Pass 'all_stopwords' variable as a parameter in set().
Then send it in for loop of iterative variable(word). Finally use stemming method(stem()). And pass the
'word' varaible as a parameter to stem().

We now join all the individual words to a single review(original state) to append all these into corpus list.
Remember, to use a space with dot parameter in join() so as to avoid the sticking of all the words without space."""

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0,1000):
    review = re.sub('[^a-zA-Z]', ' ', df['Review'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
```

# Creating Bag of Words model

We had completed the cleaning process of all the reviews. Our next step is to create the **bag of words model** in the **sentiment analysis** which will consists of **sparse matrix** with all words of all reviews in a different columns.

**Bag of Words** is the **heart of sentiment analysis**

So, we will now **create a sparse matrix that contains all the words of all the reviews in different columns** and that's the future of our **matrix of features** combined with the **dependent variable vector** that contains the binary outcomes of zeros and ones to train the future machine learning model **naive bayes algorithm to make the model understand the texts and words whether the review is bad or good (positive or negative) with the resultant outcome of 0 or 1.**

First we will create a sparse matrix. It contains rows of all the different reviews and columns of all the different words of reviews. And in each cell we will either get a 0 or 1. We will get **0 if the word of the column is not in the review of the row and get 1 if the word of the column is part of the words in the review of the row**.

```python
"""Tokenization will be done using scikit learn(sklearn) library. Specifically from a module called feature extraction
and from submodule called text and from the class CountVectorizer.

Let's import CountVectorizer class from text sub module of feature extraction module of sklearn library.

Next, we create an object for CountVectorizer class and need to pass an argument as the 'max.size of sparse matrix'.
It's nothing but the max.no. of columns. In other words, max.no. of words that we want to include in the column of 
sparse matrix. There are still some words in reviews even after cleaning which will not make sense to predict the output.
To get rid of them, we will use this parameter. And we will consider words that are most frequently appearing
in the reviews. We will not pass the value for this parameter until we train the naive bayes model on the training set.


Next, we will create the sparse matrix which is the matrix of features. We will pass 'corpus' as a parameter in
fit_transform(). and we need to convert the matrix of features into 2D array. Since when we train the naive bayes model,
it expects the matrix of features as an array.

Next, we will create a dependent variable vector. The way to get this, to take the last column of our dataset."""

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
X = cv.fit_transform(corpus).toarray()
y = df.iloc[:,-1].values
```

```python
"""Once, the above cell is executed, we need to check the no.of columns resulting the tokenization. Or the no.of words
in sparse matrix. To do this, we check the length of matrix of features of first row(X[0])."""

len(X[0])
```

```python
"""After executing the length of words, we came to know that the value is 1566. So the sparse matrix has 1566 columns.
We can round it off to 1500 and then pass this value as a parameter(max_features) in CountVectorizer class and
execute previous steps once again before proceeding next steps. After rerunning, we should get the length as 1500."""
```

# Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# Training the naive bayes model on the training set

The reason for using the naive bayes algorithm, is this model does very well with natural language processing algorithms.

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train,y_train)
```

# Predicting the test results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
